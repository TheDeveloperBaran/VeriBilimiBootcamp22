```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("C:\\Users\\carlo\\OneDrive\\Masaüstü\\Data Science x AI\\athlete_events.csv")
```

```python
data.head(1)
```

```python
data.rename(columns = {"ID": "id",
                       "Name": "isim",
                       "Sex" : "cinsiyet",
                       "Age" : "yas",
                       "Height": "boy",
                       "Team": "takim",
                       "Weight": "kilo",
                       "NOC": "kod",
                       "Games" : "oyunlar",
                       "Year": "yil",
                       "Season": "sezon",
                       "City": "sehir",
                       "Sport": "spor",
                       "Event": "etkinlik",
                       "Medal": "madalya"}, inplace = True)
```

```python
data.head()
```

```python
data = data.drop(["id", "oyunlar"], axis = 1)
```

```python
data.head(25)
```

```python
data_uniq = pd.unique(data.etkinlik)
print("Essiz etkinlik sayisi: {}".format(len(data_uniq)))
data_uniq[:15]
```

```python
data_temp = data.copy()
```

```python
boy_kilo_liste = ["boy", "kilo"]
for e in data_uniq:
    event_filter = data_temp.etkinlik == e
    data_filter = data_temp[event_filter]
    for i in boy_kilo_liste:
        ortalama = np.round(np.mean(data_filter[i]),2)
        if ~np.isnan(ortalama):
                data_filter[i] = data_filter[i].fillna(ortalama)
        else:
                all_ortalama = np.round(np.mean(data_temp[i]), 2)
                data_filter[i] = data_filter[i].fillna(all_ortalama)

    data_temp[event_filter] = data_filter

data = data_temp.copy()
data.info()
```

```python
yas_ort = np.round(np.mean(data_temp.yas), 2)
data_temp["yas"] = data_temp["yas"].fillna(yas_ort)
print("yas ortalamasi: {}".format(yas_ort))
data_temp.info()
```

```python
data = data_temp.copy()
data.info()
```

```python
data_medal = data["madalya"]
pd.isnull(data_medal).sum()
data_medal_filter = ~pd.isnull(data_medal)
data = data[data_medal_filter]
data.head()
data.info()
data.to_csv("data_cleaned.csv", index = False)
```

```python
plt.boxplot(data.yas)
plt.title("Yaş Değişkeni için Kutu Grafiği")
plt.xlabel("Age")
plt.ylabel("Value")
plt.show()
```

```python
# Çubuk (Bar, Stick) Grafiği çizdiren fonksiyon

def PlotBar(degisken, n = 5):
    data_ = data[degisken]
    data_counter = data_.value_counts()
    data_counter = data_counter[:n]
    plt.figure()
    plt.bar(data_counter.index, data_counter, color = "magenta")
    plt.xticks(data_counter.index, data_counter.index.values)
    plt.xticks(rotation = 80)
    plt.ylabel("Değişken")
    plt.title("Veri Sıklığı - {}".format(degisken))
    plt.show()
    print("{}: {}\n".format(degisken, data_counter))

categorical_var = ["isim", "cinsiyet", "sezon", "takim", "sehir", "spor"]
for i in categorical_var:
    PlotBar(i)
```
