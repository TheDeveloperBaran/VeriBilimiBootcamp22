```python
essiz_etkinlik = pd.unique(veri.olay)
```

```python
print("Eşsiz Olimpiyat Yarışı Tema Sayısı: {}".format(len(essiz_etkinlik)))
essiz_etkinlik[:3]
```

```python
veri_cpy = veri.copy()
```

```python
boy_kilo_lst = ["boy", "kilo"]
```

```python
for i in essiz_etkinlik:
    etkinlik_filtre = veri_cpy.olay == i
    veri_filt = veri_cpy[etkinlik_filtre]
    for e in boy_kilo_lst:
        print(veri_filt[e])
        ortalama = np.round(np.mean(veri_filt[e]), 2)
        print("Ortalama: {}".format(ortalama))
        if ~np.isnan(ortalama): #Ortalama varsa, etkinlikteki ortalamayla o katılımcının verilerini doldur
            veri_filt[e] = veri_filt[e].fillna(ortalama)
        else: #Ortalama yoksa o etkinlik üzerinde, o zaman tüm verideki boy kilo ortalamasını yapıştır o kişiye.
            tum_veri_ortalamasi = np.round(np.mean(veri[e]), 2)
            veri_filt[e] = veri_filt[e].fillna(tum_veri_ortalamasi)
        veri_cpy[etkinlik_filtre] = veri_filt
veri = veri_cpy.copy()
veri.info()
```
