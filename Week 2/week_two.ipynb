```python
essiz_etkinlik = pd.unique(veri.olay)
```

```python
print("Eşsiz Olimpiyat Yarışı Tema Sayısı: {}".format(len(essiz_etkinlik)))
essiz_etkinlik[:3]
```

```python
veri_cpy = veri.copy()
```

```python
boy_kilo_lst = ["boy", "kilo"]
```

```python
for i in essiz_etkinlik:
    etkinlik_filtre = (veri_cpy.olay == i)
    #print("Etkinlik filtre: \n", etkinlik_filtre)
    veri_filt = veri_cpy[etkinlik_filtre]
    #print("veri filtre: ", veri_filt)
    for e in boy_kilo_lst:
        #print("VERİ: ", veri_filt[e])
        ortalama = np.round(np.mean(veri_filt[e]), 2)
        #print("Ortalama: {}".format(ortalama))
        if ~np.isnan(ortalama): #Ortalama varsa, etkinlikteki ortalamayla o katılımcının verilerini doldur
            veri_filt[e] = veri_filt[e].fillna(ortalama)
        else: #Ortalama yoksa o etkinlik üzerinde, o zaman tüm verideki boy kilo ortalamasını yapıştır o kişiye.
            tum_veri_ortalamasi = np.round(np.mean(veri[e]), 2)
            veri_filt[e] = veri_filt[e].fillna(tum_veri_ortalamasi)
        veri_cpy[etkinlik_filtre] = veri_filt
veri = veri_cpy.copy()
veri.info()
```

```python
def plotHistogram(variable):
    plt.figure()
    plt.hist(veri[variable], bins = 91, color = "cyan")
    plt.xlabel(variable)
    plt.ylabel("Frequency in hz.")
    plt.title("Data Occurency for {} variable".format(variable))
    plt.show()

variables = ["yaş", "boy", "kilo", "yıl"]
for i in variables:
    plotHistogram(i)
```

```python
plt.boxplot(veri.yaş)
plt.title("yaş degiskeni icin kutu grafigi")
plt.xlabel("yaş")
plt.ylabel("value")
```

```python
def plotBar(degisken, n = 5):
    veri_specific = veri[degisken]
    veri_counter = veri_specific.value_counts()
    veri_counter = veri_counter[:n]
    plt.figure()
    plt.bar(veri_counter.index, veri_counter, color = "red")
    plt.xticks(veri_counter.index, veri_counter.index.values)
    plt.xticks(rotation=45)
    plt.ylabel("How Many Times?")
    plt.title("{} degiskeni icin How Many Times? bar grafigi".format(degisken))
    plt.show()
    print("{}".format(veri_counter))
```

```python
kategorik_degiskenler = ["ad", "cinsiyet", "takım", "uoc", "oyunlar", "sezon", "şehir", "spor","olay","madalya"]
for i in kategorik_degiskenler:
    plotBar(i)
```
